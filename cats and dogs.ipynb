```python
#importing libraries

import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
#loading our dataset...........
#The dataset used is cats and dogs from microsoft kaggle competition and found in net
datadir="/home/shyam/datasets/catvsdog/PetImages"
categories=['Cat','Dog']
img_size=100
```

```python
#creating a function to get our training data
train_data=[]

def create_training_data():
    for category in categories:
        path=os.path.join(datadir,category)     #path
        class_num=categories.index(category)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_arr=cv2.resize(img_array,(img_size,img_size))
                train_data.append([new_arr,class_num])
            
            except Exception as e:
                pass
        
create_training_data()
```

```python
#checking the length of the training data
print(len(train_data))
```

```python
#shuffling our data
import random

random.shuffle(train_data)
```

```python
#checking our shuffling
for sample in train_data:
    print(sample[1])
```

```python
#getting our training data
x=[]
y=[]
for features,labels in train_data:
    x.append(features)
    y.append(labels)
    
x=np.array(x).reshape(-1,img_size,img_size,1)
```

```python
#saving data using pickle
import pickle

pickle_x=open("x.pickle","wb")
pickle.dump(x,pickle_x)
pickle_x.close()

pickle_y=open("y.pickle","wb")
pickle.dump(y,pickle_y)
pickle_y.close()
```

```python
#importing libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D
```

```python
#scaling down the values
x=np.array(x/255.0)
y=np.array(y)
```

```python
#Building model
model=Sequential()

#First layer
model.add(Conv2D(64,(3,3),input_shape=x.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

#Second layer
model.add(Conv2D(64,(3,3),input_shape=x.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

#Adding a dense layer
model.add(Flatten())
model.add(Dense(64))

#output layer
model.add(Dense(1))
model.add(Activation("sigmoid"))

#compiling our model
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])

#fitting our model
model.fit(x,y,batch_size=32,validation_split=0.1,epochs=3)
```
